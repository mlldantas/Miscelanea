# Let's learn how to make plots

```julia
x = -3:0.1:3                     # creating a vector (?) with range (-3,3) and with steps of 0.1
```

```julia
typeof(x)
```

```julia
f(x) = x^2
```

```julia
y = f.(x)
```

```julia
gr()  # this is one type of backend -- whatever this means!
```

```julia
plot(x, y, label="continuous plot")
scatter!(x, y, label="scatter plot") # the "!" command serves to make the scatter plot 
                                     # overlapped with the continuous line!
```

```julia
Pkg.add("PlotlyJS")
```

```julia
plotlyjs()
```

```julia
plot(x, y, label="continuous plot")
scatter!(x, y, label="scatter plot")
```

```julia
global_temperatures = [14.4, 14.5, 14.8, 15.2, 15.5, 15.8]
npirates            = [45000, 20000, 15000, 5000, 400, 17]
```

```julia
plot(npirates, global_temperatures, legend=false)
scatter!(npirates, global_temperatures, legend=false)
xflip!()
xlabel!("Approx. number of pirates")
ylabel!("Global temperature (C)")
title!("Influence of pirate population on global warming?")
```

## Subplots in Julia

```julia
p1 = plot(x, x)
p2 = plot(x, x.^2)
p3 = plot(x, x.^3)
p4 = plot(x, x.^4)
plot(p1,p2,p3,p4, layout=(2,2), legend=false)
```
